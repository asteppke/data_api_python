```python
import data_api
api = data_api.configure()
```

```python
import datetime
time_end = datetime.datetime.now()
time_start = time_end - datetime.timedelta(minutes=1)
time_start = time_start.strftime("%Y-%m-%d %H:%M:%S")
time_end = time_end.strftime("%Y-%m-%d %H:%M:%S")
# time_start = "2016-08-03 15:45:00"
# time_end = "2016-08-03 15:55:00"
```

```python
data = api.get_data(channels=['SINSB01-RIQM-DCP10:FOR-PHASE-AVG', 
                              'SINSB01-RKLY-DCP10:FOR-PHASE-AVG'],
                    start=time_start, 
                    end=time_end, 
                    range_type="date")
```

```python
data.describe()
```

```python
data.head()
```

```python
# Persist to HDF5
data.to_hdf("test.h5", "/dataset")
```

```python
# Read from HDF5 into pandas
import pandas
data = pandas.read_hdf('test.h5')
```
